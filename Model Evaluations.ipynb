## How Model Evaluation Works 
 
 ### The 'find_model' function takes in four parameters:
     - 1. dataframe
     - 2. what features you'd like to use
     - 3. type of model
     - 4. how many classifiers
     
 #### The purpose of this function is to run a base model of the chosen classifier and print the confusion matrix, the classification report, and a summary of the scores as well as the classification results. The initial model is then saved to the 'Models' folder. The function then runs a grid search to find the optimal paramters and inserts the paramters into the model and reruns the tuned model. Another print of confusion matrix, classification report and classification results are output. The last thing that prints is the both confusion matrices as graphs. The blue graph is the non optimized model and the purple graph is the optimized model. The optimized model is then saved. 
 
 ##### How the parameters work

    1. The dataframe and classifier parameters need to match. If they do not, you will receive an error telling you that the expected and given size do not match
    2. There are three options available: 
        - all_features: a combination of the engineered features and spotify's high level analysis
        - spotify_features: spotify feature's only
        - engineered_features: engineered features only
    3. The four type of models that are currently available are:
        - KNN
        - Random Forest
        - Support Vector Classifier
        - Logistic Regression
        
    4. As mentioned, this parameter needs to match with the dataframe so the number of classifiers expected to be used is the same that the model is trained on.
